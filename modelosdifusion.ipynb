# ** Modelos de Difusión**
# Generacion de imágenes a partir de texto


```python
!pip install diffusers[torch]
!pip install transformers

from diffusers import StableDiffusionPipeline
import torch

from google.colab import drive
drive.mount('/content/drive/', force_remount = True)

import sys
sys.path.append('/content/drive/My Drive/Colab Notebooks/')
```

```python
PATH = '/content/drive/My Drive/Colab Notebooks/'
filename= 'gatovolao.jpg' #Nombre de tu imagen dentro del directorio indicado

DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
MODEL_ID = 'stabilityai/stable-diffusion-2-1'

pipeline = StableDiffusionPipeline.from_pretrained(MODEL_ID, torch_dtype=torch.float32).to(DEVICE)

input_text = 'hyperrealistic portrait of a man as astronaut, portrait, well lit, cyberpunk,'
#prompts


result = pipeline(input_text, num_inference_steps=50, guidance_scale=7).images[0]
result.save(PATH + filename)

# num_inference_steps  (iteraciones para eliminacion de ruido)
# guidance_scale controla cuánto condiciona el input_text en la imagen de salida; valores chicos tiene poco aporte en la salida
```
